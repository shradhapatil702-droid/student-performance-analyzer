# Student Performance Analyzer 

## Objective
This project explores student academic performance and identifies
factors that influence final grades (G3) using Exploratory Data Analysis (EDA).

## Dataset
- Student Performance Dataset (UCI Machine Learning Repository)
- Includes demographic, family, social, and academic attributes
- Target variable: Final Grade (G3)


```python
import pandas as pd

df=pd.read_csv(r"C:\Users\hp\Downloads\Student Performance Analyzer\data\student-data-info.txt",sep=';')
df.head()
```

```python
df.isnull().sum()  #counting missing value
```

```python
df.to_csv(r"C:\Users\hp\Downloads\Student Performance Analyzer\data\student_data_clean.csv",index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='sex', y='G3', data=df)
plt.title("Gender vs Final Grade (G3)")
plt.xlabel("Gender")
plt.ylabel("Final Grade")
plt.show()
```

### Insight
There is no significant difference in median final grades between male and female students,
though slight variations are visible across individuals.


```python
sns.boxplot(x='studytime',y='G3',data=df)
plt.title("Study Time vs Final Grade (G3)")
plt.xlabel("Weekly study time (1=<2 hrs,4=>10 hrs)")
plt.ylabel("Final grade")
plt.savefig(r"C:\Users\hp\Downloads\Student Performance Analyzer\images\studytime_vs_g3.png", bbox_inches="tight")
plt.show()
```

### Insight
Students who spend more time studying per week generally achieve higher
final grades. This suggests study time has a positive impact on academic performance.


```python
sns.scatterplot(x='absences',y='G3',data=df)
plt.title("Absences vs Final Grade (G3)")
plt.xlabel("Number of absences")
plt.ylabel("Final Grade")

plt.savefig(r"C:\Users\hp\Downloads\Student Performance Analyzer\images\absences_vs_g3.png")

plt.show()
```

### Insight
There is a slight negative trend indicating that students with more school absences
tend to have lower final grades. Regular attendance appears to positively influence academic performance.
